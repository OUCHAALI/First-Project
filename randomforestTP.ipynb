```python
import pandas as pd
from google.colab import files
upload = files.upload()
df = pd.read_csv('heart.csv')
```

```python
display(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
feuater = df.drop(columns="target")
target = df["target"]
X_train, X_test, y_train, y_test = train_test_split(feuater, target, test_size=0.20)
model = RandomForestClassifier()
model.fit(X_train,y_train)
predictions = model.predict(X_test)
```

```python
dfrs = pd.DataFrame({"Target" : y_test,"predicted_TR" : predictions})

display(dfrs)
```
